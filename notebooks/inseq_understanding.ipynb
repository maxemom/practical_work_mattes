```python
import inseq
from inseq.data import AggregatorPipeline

model = inseq.load_model("Qwen/Qwen2-1.5B", "saliency")
out = model.attribute(
  "Samantha saw a boy at school that she thought was cute. Samantha was shy about asking the boy out. She decided to ask the boy out anyway. The boy said yes because he liked Samantha. They are going on a date this Friday! Why did Samantha see a boy?", 
  generation_args={"max_new_tokens": 1}
)
out.show(do_aggregation=True)
```

```python
import torch, inseq

model = inseq.load_model("Qwen/Qwen2-1.5B", "saliency")
p = next(model.model.parameters())

print("device:", p.device)
print("dtype :", p.dtype)
print("MPS available:", torch.backends.mps.is_available())
```

```python
model.model = model.model.to("mps")
p = next(model.model.parameters())
print("device forced:", p.device)
```

```python
out = model.attribute(
    "Test prompt",
    generation_args={"max_new_tokens": 3}
)
```

```python
seq = out.sequence_attributions[0]
raw_source = seq.source_attributions
raw_target = seq.target_attributions
print(raw_target.shape)
```
