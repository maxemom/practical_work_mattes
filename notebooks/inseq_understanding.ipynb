```python
import torch
import inseq

device = "mps" if torch.backends.mps.is_available() else "cpu"

model = inseq.load_model(
    "Qwen/Qwen2.5-0.5B-Instruct",
    "saliency",
    device=device,
    model_kwargs={
        "torch_dtype": torch.float16
    }
)

out = model.attribute(
    input_texts="The capital of France is",
    generation_args={
        "max_new_tokens": 2
    }
)

out.show()
```

```python
import inseq
import torch

device = "mps" if torch.backends.mps.is_available() else "cpu"

model = inseq.load_model(
    "Qwen/Qwen2.5-0.5B-Instruct",
    "saliency",
    device=device
)

prompt = "The capital of France is"
gen_token_ids = [12095, 13]

# Nur die neue Fortsetzung decodieren
continuation = model.tokenizer.decode(gen_token_ids, skip_special_tokens=False)
print("Continuation repr:", repr(continuation))

# Für decoder-only: voller Zieltext = Prompt + Fortsetzung
full_generated_text = prompt + continuation
print("Full generated text repr:", repr(full_generated_text))

out = model.attribute(
    input_texts=prompt,
    generated_texts=full_generated_text
)

out.show()
```
